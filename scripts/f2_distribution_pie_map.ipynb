```python
from matplotlib import pyplot as plt
import matplotlib as mpl

mpl.rcParams['font.size'] = 18
plt.figure(figsize=(12,10))#调节图形大小

ax1 = plt.subplot(221)
ax2 = plt.subplot(222)
ax3 = plt.subplot(224)
ax4 = plt.subplot(223)
plt.subplots_adjust(wspace =1.5, hspace =0)
labels1 = [u'oceans',u'other \n storages'] #定义标签
sizes1 = [96.5,3.5]#每块值
colors1 = ['#60ACFC','#FF7C7C'] #每块颜色定义
labels2 = [u'glaciers and icecaps', u'groundwater', u'other \n storages']
sizes2 = [1.74,1.7,0.08]
colors2 = ['#32D3EB','#FEB64D','#9287E7']

labels3 = [u'lakes', u'permafrost', u'other \n storages']
sizes3 = [13,22,0.31]
colors3 = ['#5BC49F','#9287E7','#60ACFC']

labels4 = [u'atmosphere', u'soil moisture', u'river and swamp', u'biological \n water']
sizes4 = [1,1,1,0.1]
colors4 = ['#FF7C7C','#60ACFC','#5BC49F','#FEB64D']
explode1 = (0, 0.2) 
explode2 = (0,0.02,0.2)
explode3 = (0,0.02,0.2)
explode4 = (0.05,0.05,0.05,0.05)
#将某一块分割出来，值越大分割出的间隙越大
ax1,text1,text2 = ax1.pie(sizes1,explode=explode1,
                      labels=labels1,
                      colors=colors1,
                      autopct = '%3.2f%%', #数值保留固定小数位
                      shadow = True, #无阴影设置
                      startangle =0, #逆时针起始角度设置
                      pctdistance = 0.6) #数值距圆心半径倍数的距离)
ax2,text3,text4 = ax2.pie(sizes2,explode=explode2,
                      labels=labels2,
                      colors=colors2,
                      autopct = '%3.2f%%', #数值保留固定小数位
                      shadow = True, #无阴影设置
                      startangle =300, #逆时针起始角度设置
                      pctdistance = 0.6) #数值距圆心半径倍数的距离)
ax3,text5,text6 = ax3.pie(sizes3,explode=explode3,
                      labels=labels3,
                      colors=colors3,
                      autopct = '%3.2f%%', #数值保留固定小数位
                      shadow = True, #无阴影设置
                      startangle =175, #逆时针起始角度设置
                      pctdistance = 0.6) #数值距圆心半径倍数的距离)
ax4,text7,text8 = ax4.pie(sizes4,explode=explode4,
                      labels=labels4,
                      colors=colors4,
                      autopct = '%3.2f%%', #数值保留固定小数位
                      shadow = True, #无阴影设置
                      startangle =175, #逆时针起始角度设置
                      pctdistance = 0.6#数值距圆心半径倍数的距离)
                      ) 


#plt.axis('equal')

plt.savefig('water_distribution.jpg')
plt.show()
```
